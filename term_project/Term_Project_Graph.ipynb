```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import kde
```

```python
data1 = pd.read_csv('OutputALL/part-00000', sep="\t", header=None)
data2 = pd.read_csv('OutputALL/part-00001', sep="\t", header=None)
data = data1.append(data2)
x = []
y =[]
for index, row in data.iterrows():
    if(len(row.values[0][1:-1].split(", ")) == 3):
        x.append(float(row.values[0][1:-1].split(", ")[1]))
        y.append(float(row.values[0][1:-1].split(", ")[2]))
    if(len(row.values[0][1:-1].split(", ")) == 4):
        x.append(float(row.values[0][1:-1].split(", ")[2]))
        y.append(float(row.values[0][1:-1].split(", ")[3]))
    if(len(row.values[0][1:-1].split(", ")) == 5):
        x.append(float(row.values[0][1:-1].split(", ")[3]))
        y.append(float(row.values[0][1:-1].split(", ")[4]))
```

```python
plt.figure(figsize=(20,10))
plt.scatter(x, y, alpha=0.3)
plt.xlabel('Zipcode')
plt.ylabel('Rating')
plt.show()
# plt.xticks([])
# plt.yticks([])
```

```python
data1 = pd.read_csv('OutputGQC/part-00000', sep="\t", header=None)
data2 = pd.read_csv('OutputGQC/part-00001', sep="\t", header=None)
data = data1.append(data2)
x = []
y =[]
xCentroid = []
yCentroid = []
plt.figure(figsize=(20,10))
for index, row in data.iterrows():
    xCentroid.append(float(row.values[0][1:-1].split(",[")[0][1:-1].split(", ")[1]))
    yCentroid.append(float(row.values[0][1:-1].split(",[")[0][1:-1].split(", ")[2]))
    x = []
    y = []
    for z in row.values[0][1:-1].split(",[")[1][1:-2].split("), ("):
        if(len(z.split(", ")) == 3):
            x.append(float(z.split(", ")[1]))
            y.append(float(z.split(", ")[2]))
        if(len(z.split(", ")) == 4):
            x.append(float(z.split(", ")[2]))
            y.append(float(z.split(", ")[3]))
        if(len(z.split(", ")) == 5):
            x.append(float(z.split(", ")[3]))
            y.append(float(z.split(", ")[4]))
    
    plt.scatter(x, y, alpha=0.3)
plt.scatter(xCentroid, yCentroid, marker='x', c='k')
plt.show()
```
